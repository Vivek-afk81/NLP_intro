```python
text="this is a classic Anaconda environment mismatch problem"

text=text.lower()

print("Lower text: ",text)
```

##step 2 :-Remove Punctuations
we we will remove everything except letters and spaces
Using re- regular expression

```python
import re
text =re.sub(r'[^\w\s]','',text)
print("Without punctuation: ",text)

#[^\w\s] - remove everything except letters and spaces
```

##STEP-3 TOKENIZATION

```python
tokens=text.split()
print("tokens",tokens)
```

##STEP-4 Remove Stop Words

```python
#remove stop words is am are the etc

stopwords = [
    "a", "an", "the",
    "is", "am", "are", "was", "were",
    "this", "that", "these", "those",
    "and", "or", "but",
    "to", "of", "in", "on"
]
filtered_tokens=[]

for word in tokens:
  if word not in stopwords:
    filtered_tokens.append(word)

print(filtered_tokens)
```

```python
def stem(word):
  if word.endswith("ing"):
    return word[:-3]
  return word

stemmed_words=[]
for word in filtered_tokens:
  stemmed_words.append(stem(word))

print("After stemming :",stemmed_words)
```
